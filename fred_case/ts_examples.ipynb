Description of variables:
- GS10 - 10-Year Treasury Constant Maturity Rate (monthly)
- EXCHUS - China US Foreign Exchange Rate (monthly)
- EXCAUS - Canadian to US Dollar Exchange Rate Daily (monthly)
- Weekly U.S. Refiner and Blender Adjusted Net Production of Finished Motor Gasoline (Thousand Barrels per Day) (weekly)
- Weekly Minnesota Midgrade Conventional Retail Gasoline Prices (Dollars per Gallon) (weekly)
- Weekly U.S. Percent Utilization of Refinery Operable Capacity (Percent) (weekly)
- Weekly U.S. Exports of Crude Oil and Petroleum Products (Thousand Barrels per Day) (weekly)
- Weekly U.S. Field Production of Crude Oil (Thousand Barrels per Day) (weekly)
- Weekly U.S. Ending Stocks of Crude Oil and Petroleum Products (Thousand Barrels) (weekly)
- Weekly U.S. Product Supplied of Finished Motor Gasoline (Thousand Barrels per Day) (weekly)
- temp - temperature (hourly)
- traffic_volume - Traffic data from the MN DOT via the UCI data repository (hourly)

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 11, 4
```

```python
short_df = pd.read_csv('data/ts_short.csv')
short_df['datetime'] = pd.to_datetime(short_df['datetime'])
long_df = pd.read_csv('data/ts_long.csv')
long_df['datetime'] = pd.to_datetime(long_df['datetime'])
```

```python
for label in short_df['series_id'].unique():
    df = short_df[short_df['series_id'] == label]
    
    print(f'Length of the time series: {len(df)}')
    plt.plot(df['datetime'], df['value'])
    plt.title(label)
    plt.grid()
    plt.show()
```

```python
for label in long_df['series_id'].unique():
    df = long_df[long_df['series_id'] == label]
    
    print(f'Length of the time series: {len(df)}')
    plt.plot(df['datetime'], df['value'])
    plt.title(label)
    plt.grid()
    plt.show()
```
